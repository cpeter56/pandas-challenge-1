## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
subtotal = df['unit_price'] * df['qty']
df["Subtotal"] = subtotal
df
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
total_weight = df['unit_weight'] * df['qty']
df["Total Weight"] = total_weight

def shipping_price(row):
    if row["Total Weight"] > 50:
        return row["Total Weight"] * 7
    else:
        return row["Total Weight"] * 10
    
df["Shipping Price (LB's)"] = df.apply(shipping_price, axis=1)
df
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
Total_Price1 = df["Subtotal"] + df["Shipping Price (LB's)"]

Total_Price = (Total_Price1 * 0.0925) + Total_Price1
df["Total Price"] = Total_Price.round(2)
df
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
line_subtotal = df["unit_cost"] * df["qty"]

df["Line Subtotal"] = line_subtotal

line_cost = line_subtotal + df["Shipping Price (LB's)"]

df["Line Cost"] = line_cost

df
```

```python
# Create a column for the profit of each line using line cost and line price
profit = df["Total Price"] - df["Line Cost"]

df["Profit"] = profit

df
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
for client_id in [33615, 66037, 46820, 38378, 24741]:
    Total_Spend = df[df['client_id'] == client_id]['Total Price'].sum()
    print(f"For the Client ID {client_id}, the Total Spend is: ${Total_Spend:.2f}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
top_5_clients = []
for client_id in [33615, 66037, 46820, 38378, 24741]:
    Total_Quantity = df[df['client_id'] == client_id]['qty'].sum()
    Total_Shipping = df[df['client_id'] == client_id]['Shipping Price (LB\'s)'].sum()
    Total_Price = df[df['client_id'] == client_id]['Total Price'].sum()
    Total_Line_Cost = df[df['client_id'] == client_id]['Line Cost'].sum()
    Total_Profit = df[df['client_id'] == client_id]['Profit'].sum()

    top_5_clients.append({
        'client_id': client_id,
        'Total_Quantity': Total_Quantity,
        'Total_Shipping': Total_Shipping,
        'Total_Price': Total_Price,
        'Total_Line_Cost': Total_Line_Cost,
        'Total_Profit': Total_Profit
    })

#top_5_clients_df = df.loc[df['client_id'].isin([33615, 66037, 46820, 38378, 24741]), ['client_id', 'qty', 'Shipping Price (LB\'s)', 'Total Price', 'Line Cost', 'Profit']] 

top_5_clients_df = pd.DataFrame(top_5_clients)

top_5_clients_df
```

```python
# Format the data and rename the columns to names suitable for presentation.
top_5_clients_df_renamed = top_5_clients_df.rename(columns={"client_id": "Client ID", 
                                                            "Total_Quantity": "Units", 
                                                            "Total_Shipping": "Shipping (millions)",
                                                            "Total_Price": "Total Revenue (millions)",
                                                            "Total_Line_Cost": "Total Cost (millions)",
                                                            "Total_Profit": "Total Profit (millions)"
    
})
# Define the money columns. 
money_columns = ['Shipping (millions)', 'Total Revenue (millions)', 'Total Cost (millions)', 'Total Profit (millions)']
# Define a function that converts a dollar amount to millions.
def currency_format_millions(x):
    return x/1000000

   

# Apply the currency_format_millions function to only the money columns. 
top_5_clients_df_renamed[money_columns] = top_5_clients_df_renamed[money_columns].applymap(currency_format_millions)

# Rename the columns to reflect the change in the money format. 
top_5_clients_df_renamed

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
total_profit_sorted = top_5_clients_df_renamed.sort_values(by=['Total Profit (millions)'], ascending=False)
total_profit_sorted
```
